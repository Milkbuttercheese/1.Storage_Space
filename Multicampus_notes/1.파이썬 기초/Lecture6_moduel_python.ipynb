```python
'''
try, except 구문
'''
a=5 
b=0
try :
    c= a/b
    print(c)
except (ZeroDivisionError,NameError):
    print('ZeroDivisonError가 발생했습니다\n')
    print('c에 0을 저장합니다')
    c=0
    print(c)
```

```python
'''
정상 수행 문장에는 spam()함수를 호출
NameError가 발생한 경우 예외처리- as x
x 출력해보기
'''
try: 
    spam()
except NameError as x:
    print(x)
```

```python
'''
정상 수행 문장에는 spam()함수를 호출
NameError가 발생한 경우 예외처리- as x
x 출력해보기
무적권 
'''
try: 
    spam()
except NameError as x:
    print(x)
finally:
    print('스팸은 맛있어')
```

```python
'''
MyError 클래스 생성하기(Exception 클래스에서 상속받기)
아무 기능 수행 X
예외 직접 만들기
'''
class MyError(Exception):
    pass

#raise는 내가 정의한 클래스를 에러로 표현할 수 있다
def say_nick(nick):
    if nick =='바보':
        raise MyError()
    print(nick)
    
# print(say_nick('천사'))
print(say_nick('바보'))
```

```python
'''
MyError 클래스 생성하기(Exception 클래스에서 상속받기)
Exception은 Error를 담고있는 내장된 클래스
아무 기능 수행 X
예외 직접 만들기

__str__ : 클래스 자체의 정보를 출력하는 명령어

'''

class MyError(Exception):
    def __str__(self):
        return '허용되지 않는 별명입니다.'

#raise는 내가 정의한 클래스를 에러로 표현할 수 있다
def say_nick(nick):
    if nick =='바보':
        raise MyError()
    print(nick)
    
# print(say_nick('천사'))
print(say_nick('바보'))
    
    
```

```python
'''
filter 함수 사용 전
'''

def positive(x):
    result= []
    for i in x:
        if i>0:
            result.append(i)
    return result

x=[1,2,-3,-5,6,-7,20,-15,3]
print(positive(x))
```

```python
'''
리스트에 부등호 사용? 될리가 없지
'''

def positive_2(x):
    return x>0

x=[1,2,-3,-5,6,-7,20,-15,3]
print(positive_2(x))
```

```python
'''
filter 함수
인자 2개 필요( 필터 역할 함수, 필터에 걸러질 자료형)
'''
def positive_2(x):
    return x>0

x=[1,2,-3,6,12,-24]
print(positive(x))

a=list(filter(positive_2,[1,2,-3,6,12,-24]))
print(a)
```
