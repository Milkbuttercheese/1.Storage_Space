```python
df= pd.read_excel('시도별 전출입 인구수.xlsx',header=0)
#서울 -> 다른 지역 이동
#1. fillna 함수를 사용하여 NaN을 체워주기option: method='ffill' : 위에 있는 데이터를
df=df.fillna(method='ffill')
df
#서울에서 다른 지역으로 이동한 데이터 추출
df_seoul= df[(df['전출지별'] == '서울특별시') & (df['전입지별']!= '서울특별시')]

#전출지별 삭제
#전입지별-> 전입지로 변경
#전입지를 인덱스로 변경
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
dt_g=df_seoul.loc['경기도']
#스타일 서식 지정
plt.style.use('ggplot')
```

```python
#화면 분할을 하되 1행 1열짜리 화면을 만들기
fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(1,1,1)
ax.plot(dt_g,marker='o',markerfacecolor='green',markersize=10,color='olive',
         linewidth=2,label='서울-> 경기')
ax.legend(loc='best')
#ax y축 범위 지정
ax.set_ylim(50000,800000)
#ax 눈금 라벨 회전
ax.set_xticklabels(dt_g.index,rotation=90)

#차트제목 추가
ax.set_title('서울->경기 이동 인구',size=20)
#축 이름 정하기
ax.set_xlabel('기간',size=12)
ax.set_ylabel('이동 인구수',size=12)
#x,y축 눈금 라벨 크기 변경
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('./auto-mpg.csv',header=None)
col_names=['mpg','cylinders','displacement','horsepower','weight',
           'acceleration','model years','origin','name']
df.columns=col_names

#cylinders의 상대적 비율계산
cylinders_size=df.cylinders/df.cylinders.max() *100

df.plot(x=['mpg'],y=['weight'],marker='+',s=50,cmap='viridis',c=cylinders_size,kind='scatter',figsize=(10,5),alpha=0.3)
plt.xlabel('mpg')
plt.title('산점도')
plt.savefig('./산점도')
plt.show()
```
