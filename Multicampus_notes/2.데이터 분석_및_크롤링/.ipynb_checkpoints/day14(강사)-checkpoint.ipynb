박스 플롯
- axe 객체를 분할하여 2개의 그래프 제작
- boxplot 함수 사용
 - 수직 박스 플롯 : vert = True
 - 수평 박스 플롯 : vert = False

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')

df = pd.read_csv('./auto-mpg.csv', header = None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
              'acceleration', 'model year', 'origin', 'name']

# 그림 화면 2분할(ax1, ax2)
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 제조국가별 mpg 현황에 대한 박스 플롯(3개)
# axe 객체에 boxplot 함수 사용
# 제조국가 별로 맞는 번호에 맞춰 mpg 열의 값을 찾음
# 인자(x = 조건, labels = [국가 이름], 박스플롯 종류(vert))
# - 국가 이름 : 'USA', 'EU', 'JAPAN'
ax1.boxplot(x = [df[df['origin'] == 1]['mpg'],
                 df[df['origin'] == 2]['mpg'],
                 df[df['origin'] == 3]['mpg']],
            labels = ['USA', 'EU', 'JAPAN'],
            vert = True)

ax2.boxplot(x = [df[df['origin'] == 1]['mpg'],
                 df[df['origin'] == 2]['mpg'],
                 df[df['origin'] == 3]['mpg']],
            labels = ['USA', 'EU', 'JAPAN'],
            vert = False)

plt.show()
```

회귀선이 있는 산점도
- 회귀선 : 점 간의 관계를 파악하여 그려놓은 예측 가능한 선

```python
# matplotlib 의 pyplot 불러오기(plt)
# seaborn 불러오기(sns)
import matplotlib.pyplot as plt
import seaborn as sns

# titanic 데이터 셋 가져오기(titanic)
titanic = sns.load_dataset('titanic')

# 스타일 테마(darkgrid, dark, whitegrid, white, ticks)
# seaborn.set_style(스타일 명)
sns.set_style('darkgrid')

# 화면 분할(2분할)
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 회귀선이 있는 산점도
# seaborn 모듈의 regplot 함수 사용
# 회귀선 표시 : fit_reg = True
# 회귀선 표시 X : fit_reg = False
# 인자 사용(x축, y축, data = 데이터셋 명, ax = 분할된 axe 객체, 회귀선 표시 여부)
# x축 : age
# y축 : fare
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = ax1,
            fit_reg = True)

# 회귀선 표시 X
sns.regplot(x = 'age',
            y = 'fare',
            data = titanic,
            ax = ax2,
            fit_reg = False)

plt.show()
```

히트맵
- x축 항목과 y축 항목에 해당하는 개수를 세서 값을 측정해놓은 맵

```python
# matplotlib 의 pyplot 불러오기(plt)
# seaborn 불러오기(sns)
import matplotlib.pyplot as plt
import seaborn as sns

# titanic 데이터 셋 가져오기(titanic)
titanic = sns.load_dataset('titanic')

# 스타일 테마(darkgrid, dark, whitegrid, white, ticks)
# seaborn.set_style(스타일 명)
sns.set_style('darkgrid')

# 집계
# table 객체 생성
# 데이터 셋에 pivot_table 함수 사용
# 인자(index, columns, aggfunc)
# x축에 'sex'
# y축에 'class'
# 집계에 'size'
table = titanic.pivot_table(index = 'sex', columns = 'class', aggfunc = 'size')

# seaborn 모듈의 heatmap 함수 사용
# 인자(데이터프레임, 데이터 값의 표시 여부, 포맷, 컬러맵, 구분선 굵기, 컬러바 표시 여부)
# annot = True
# fmt = d
# cmap = YlGnBu
# linewidth = .5
# cbar = True or False
sns.heatmap(table,
            annot = True, fmt = 'd',
            cmap = 'YlGnBu',
            linewidth = .5,
            cbar = False)

plt.show()
```

범주형 데이터의 산점도

```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')

sns.set_style('ticks')

# 화면 2분할
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# seaborn 객체에서 stripplot 함수 사용
# 인자(x축, y축, data = 데이터 셋, ax = 분할된 axe 객체)
# x축 : class
# y축 : age
sns.stripplot(x = "class", y = "age", data = titanic, ax = ax1)

# seaborn 객체에서 swarmplot 함수 사용
# 인자(x축, y축, data = 데이터 셋, ax = 분할된 axe 객체)
# x축 : class
# y축 : age
sns.swarmplot(x = "class", y = "age", data = titanic, ax = ax2)

# 차트 제목
# 1번 차트 제목(Strip Plot)
# 2번 차트 제목(Swarm Plot)
ax1.set_title("Strip Plot")
ax2.set_title("Swarm Plot")

plt.show()
```

지도에 마커 표시

```python
# folium 모듈 불러오기
# pandas 모듈 불러오기(pd)
import folium
import pandas as pd

# 엑셀 파일 불러오기(서울지역 대학교 위치.xlsx)
# 행 인덱스 지정 X
df = pd.read_excel('./서울지역 대학교 위치.xlsx', index_col = 0)

# Stamen Terrain 스타일 지정하여 seoul_map 객체 생성
seoul_map = folium.Map(location = [37.55, 126.98], zoom_start = 12, 
                       tiles = 'Stamen Terrain')

# for문 사용
# 데이터프레임에서 가져올 변수 생성 name, lat, lng
# 가져올 데이터 : zip(df의 인덱스, df의 위도, df의 경도)
# 실행 문장
# folium 모듈의 Marker 함수 사용
# 인자([위도, 경도], popup)
# Marker 뒤에 add_to 함수 사용
# 인자(지도 객체)
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.Marker([lat, lng], popup = name).add_to(seoul_map)

# 지도 저장(seoul_colleges.html)
seoul_map.save('./seoul_colleges.html')
```

```python
import folium
import pandas as pd

df = pd.read_excel('./서울지역 대학교 위치.xlsx', index_col = 0)

seoul_map = folium.Map(location = [37.55, 126.98], zoom_start = 12, 
                       tiles = 'Stamen Terrain')

# for문 사용
# 데이터프레임에서 가져올 변수 생성 name, lat, lng
# 가져올 데이터 : zip(df의 인덱스, df의 위도, df의 경도)
# 실행 문장
# folium 모듈의 CircleMarker 함수 사용
# 인자([위도, 경도], 이름, 반지름, 원의 둘레 색상, 원 색상을 채울 것인지의 여부,
# 원을 채우는 색상, 투명도)
# radius = 10
# color = 갈색
# fill = True or False
# fill_color = 코랄
# fill_opacity = 0.7
# Marker 뒤에 add_to 함수 사용
# 인자(지도 객체)
for name, lat, lng in zip(df.index, df.위도, df.경도):
    folium.CircleMarker([lat, lng], 
                        popup = name, 
                        radius = 10,
                        color = 'brown',
                        fill = True,
                        fill_color = 'coral',
                        fill_opacity = 0.7).add_to(seoul_map)

# 지도 저장(seoul_colleges.html)
seoul_map.save('./seoul_colleges2.html')
```

단계 구분도

```python
import folium
import pandas as pd

# json 모듈 불러오기
import json

# 경기도인구데이터 파일을 불러와서 데이터프레임으로 저장(인덱스 : '구분')
# 데이터프레임의 열을 map 함수를 사용하여 문자열 형식으로 지정
df = pd.read_excel('./경기도인구데이터.xlsx', index_col = '구분')
df.columns = df.columns.map(str)

# geo_data 객체 생성
# json 모듈에서 load 함수 사용
# 인자(open 함수 사용)
# 인자(json 파일 경로, encoding 방식)
# json 파일명 : 경기도행정구역경계.json
# encoding 방식 : utf-8
geo_data = json.load(open('./경기도행정구역경계.json', encoding = 'utf-8'))

# 지도 만들기
# g_map 객체 생성
# 줌 : 9
g_map = folium.Map(location = [37.55, 126.98], zoom_start = 9, 
                   tiles = 'Stamen Terrain')

year = '2017'

# 단계 구분도 제작
# folium 모듈의 choropleth 함수 사용
# 인자(geo_data, 데이터, columns, 채울 색, 투명도, 구분선 투명도, 범위 지정, 
# key_on = feature.properties.name) # geometry.type 등등
# data = df['2007']
# columns = [df.index, df['2007']]
# fill_color = YlOrRd
# fill_opacity = 0.7
# line_opacity = 0.3
# threshold_scale = [10000, 100000, 300000, 500000, 700000]
# key_on = feature.properties.name
# add_to 함수 사용
# 인자(지도 객체)
folium.Choropleth(geo_data = geo_data,
                  data = df[year],
                  columns = [df.index, df[year]],
                  fill_color = 'YlOrRd', fill_opacity = 0.7, line_opacity = 0.3,
                  threshold_scale = [10000, 100000, 300000, 500000, 700000], 
                  key_on = "feature.properties.name").add_to(g_map)

# g_map 저장(gyonggi_population_2007.html)
g_map.save('./gyonggi_population_' + year + '.html')
```
