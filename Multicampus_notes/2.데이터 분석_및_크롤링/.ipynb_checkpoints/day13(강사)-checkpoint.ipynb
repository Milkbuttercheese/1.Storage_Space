```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = './malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_excel("./  .xlsx", header = 0)

df = df.fillna(method='ffill')

mask = (df[''] == '') & (df[''] != '')

df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis = 1)
df_seoul.rename({'':''}, axis = 1, inplace=True)
df_seoul.set_index('', inplace=True)

dt_g = df_seoul.loc['']

plt.style.use('ggplot')

fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

# ax1    
# plot  
# ( ,  ,  ) - dt_g, 'o', markersize = 10
ax1.plot(dt_g, 'o', markersize = 10)

# ax2    
# plot  
# ( ,  ,  (),  ,  (),  , )
# - dt_g, marker = 'o', markerfacecolor = 'green', markersize = 10,
# - color = 'olive', linewidth = 2, label = ' -> '
ax2.plot(dt_g, marker = 'o', markerfacecolor = 'green', markersize = 10,
         color = 'olive', linewidth = 2, label = ' -> ')

# ax2    
# legend  
# () - loc = 'best'
ax2.legend(loc = 'best')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = './malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_excel("./  .xlsx", header = 0)

df = df.fillna(method='ffill')

mask = (df[''] == '') & (df[''] != '')

df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis = 1)
df_seoul.rename({'':''}, axis = 1, inplace=True)
df_seoul.set_index('', inplace=True)

dt_g = df_seoul.loc['']

plt.style.use('ggplot')

fig = plt.figure(figsize = (10, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(dt_g, 'o', markersize = 10)

ax2.plot(dt_g, marker = 'o', markerfacecolor = 'green', markersize = 10,
         color = 'olive', linewidth = 2, label = ' -> ')

ax2.legend(loc = 'best')

# ax1, ax2  y   
# set_ylim  
# (, ) - (50000, 800000)
ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

# ax1, ax2 x   
# set_xticklabels  
# ( .index, rotation =  )
ax1.set_xticklabels(dt_g.index, rotation = 90)
ax2.set_xticklabels(dt_g.index, rotation = 90)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = './malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_excel("./  .xlsx", header = 0)

df = df.fillna(method='ffill')

mask = (df[''] == '') & (df[''] != '')

df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis = 1)
df_seoul.rename({'':''}, axis = 1, inplace=True)
df_seoul.set_index('', inplace=True)

col_years = list(map(str, range(2010, 2018)))
df_4 = df_seoul.loc[['', '', '', ''], col_years]

# 2010 2017       
df_4[''] = df_4.sum(axis = 1)

#     
df_total = df_4[['']].sort_values(by='', ascending = True)
print(df_total)

df_total.plot(kind = 'barh', figsize = (20, 10), width = 0.7, color = 'orange')

plt.title(' ->   (2010-2017)', size = 30)

plt.ylabel(' ', size = 20)

plt.show()
```
