```python
df= pd.read_excel('시도별 전출입 인구수.xlsx',header=0)
df
```

```python
#서울 -> 다른 지역 이동
#1. fillna 함수를 사용하여 NaN을 체워주기option: method='ffill' : 위에 있는 데이터를
df=df.fillna(method='ffill')
df
#서울에서 다른 지역으로 이동한 데이터 추출
df_seoul= df[(df['전출지별'] == '서울특별시') & (df['전입지별']!= '서울특별시')]

#전출지별 삭제
#전입지별-> 전입지로 변경
#전입지를 인덱스로 변경
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace=True)
df_seoul.set_index('전입지', inplace=True)

df_seoul
```

```python
'''차트 제목과 축 이름 추가하기

'''
dt_g=df_seoul.loc['경기도']
#스타일 서식 지정
plt.style.use('ggplot')
# 그림 크기 키우기
plt.figure(figsize=(14,5))
#눈금 라벨 회전시키기
plt.xticks(rotation=90)
#글자 크기 변경
plt.title('서울-> 경기 인구 이동',size=30)
plt.xlabel('기간',size=20)
plt.ylabel('이동 인구수',size=20)
plt.legend(labels=['서울=> 경기 인구 이동'],loc='best', fontsize=15)
plt.plot(dt_g.index,dt_g.values,marker='o',markersize=10)
#y축 범위 늘리기
plt.ylim(50000,800000)
#화살표 만들기
plt.annotate('',xy=(20,620000),xytext=(2,290000),xycoords='data',
             arrowprops = dict(arrowstyle = '->', color = 'skyblue', lw = 5))
plt.annotate('',xy=(40,500000),xytext=(26,650000),xycoords='data',
             arrowprops=dict(arrowstyle='->',color='skyblue',lw=5))
plt.show()
```
