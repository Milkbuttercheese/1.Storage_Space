```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300
#plt.rcParams['figure.figsize'] = [5.0, 5.0]
#sns.set(rc = {'figure.figsize':(5,5)})
#plt.figure(dpi=300)
```

```python
#visual 01
car_crashes = sns.load_dataset('car_crashes')
car_crashes

#kaggle;https://www.kaggle.com/datasets/fivethirtyeight/fivethirtyeight-bad-drivers-dataset
#에서 columns_name에 대한 정보를 찾아보기
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=car_crashes, x='abbrev',y='total')
plt.show()
```

```python
#visual 02
car_crashes.sort_values('total',ascending=False,inplace=True)
plt.figure(figsize=(10,5))
sns.barplot(data=car_crashes,x='abbrev',y='total',facecolor='w',edgecolor='black')

sns.lineplot(data=car_crashes,x='abbrev',y='speeding',linewidth=3,color='r',label='speeding')
sns.lineplot(data=car_crashes,x='abbrev',y='alcohol',linewidth=3,color='g',label='alcohol')
sns.lineplot(data=car_crashes,x='abbrev',y='no_previous',linewidth=3,color='b',label='no_previous')
plt.xlim(-1,51)
plt.show()
```

```python
#visual 02
car_crashes.sort_values('total',ascending=False,inplace=True)
plt.figure(figsize=(10,5))
sns.barplot(data=car_crashes,x='abbrev',y='total',facecolor='w',edgecolor='black')

sns.barplot(data=car_crashes,x='abbrev',y='speeding',linewidth=3,color='r',label='speeding',alpha=0.3)
sns.barplot(data=car_crashes,x='abbrev',y='alcohol',linewidth=3,color='g',label='alcohol',alpha=0.3)
sns.barplot(data=car_crashes,x='abbrev',y='no_previous',linewidth=3,color='b',label='no_previous',alpha=0.3)
plt.xlim(-1,51)
plt.show()
```

```python
#visual 03
penguins= sns.load_dataset('penguins')
penguins
sns.histplot(data=penguins,x='body_mass_g',hue='species',multiple='fill',fill=False)
plt.show()
```

```python
sns.histplot(data=penguins,x='body_mass_g',hue='species')
plt.show()
```

```python
#visual 04
penguins= sns.load_dataset('penguins')
sns.boxplot(data=penguins,x='species',y='bill_depth_mm',hue='sex')
plt.show()
```

```python
#visual 05
penguins= sns.load_dataset('penguins')
sns.swarmplot(data=penguins, x='body_mass_g',y='species',alpha=0.5)
plt.show()
```

```python
penguins= sns.load_dataset('penguins')
markers={'Male':'P','Female':'o'}
sns.scatterplot(data=penguins,x='bill_length_mm',y='bill_depth_mm',hue='species',style='sex')
plt.show()
```

```python
sns.violinplot(data=penguins,x='body_mass_g',y='species',hue='sex',split=True)
plt.show()
```

```python
sns.regplot(data=penguins, x='bill_length_mm',y='bill_depth_mm')
plt.show()
```

```python
sns.displot(data=penguins,y='body_mass_g',hue='sex',multiple='stack',col='species')
plt.savefig('displot')
plt.show()
```
