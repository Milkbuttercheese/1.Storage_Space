```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from random import randint
```

```python
#visual 01
fig,ax= plt.subplots()

x=[randint(0,10) for i in range(100)]
ax.hist(x,bins=10)

#축 눈금
plt.xticks(list(range(0,11)))
plt.yticks(list(range(0,101,5)))

#축 제한
plt.xlim(0,10)
plt.ylim(0,100)
```

```python
#visual 02
fig,ax= plt.subplots()
x=[randint(0,10) for i in range(100)]
#cumulative
ax.hist(x,bins=10, cumulative=True)
ax.hist(x,bins=10, cumulative=False)

plt.xticks(list(range(1,11)))
plt.yticks(list(range(0,101,5)))
plt.xlim(0,10)
plt.ylim(0,100)
plt.show()
```

```python
#visual 03
fig,ax= plt.subplots()
x=list(randint(0,1000) for i in range(10))
print(f'평균{np.mean(x)}')
print(f'중위값{np.median(x)}')

ax.boxplot(x)
plt.show()
```

```python
#visual 04
fig,ax =plt.subplots()
ages= [randint(1,100) for i in range(100)]
child=[x for x in ages if x<19]
young=[x for x in ages if 19<=x<40]
middle=[x for x in ages if 40<=x<60]
old=[x for x in ages if 60<x]


labels=['child','young','middle','old']
count=[len(child),len(young),len(middle),len(old)]
ax.pie(count,labels=labels,startangle=90,counterclock=False)
plt.show()
```

```python
fig= plt.figure()
ax01=fig.add_subplot(1,2,1)
ax02=fig.add_subplot(1,2,2)

x=[1,2,3,4,5]
y=[randint(1,100) for i in range(5)]
ax01.bar(x,y,color='r')
ax02.barh(x,y,color='b')
plt.show()
```
