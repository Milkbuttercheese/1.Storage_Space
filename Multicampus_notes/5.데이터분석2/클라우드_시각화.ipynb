```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import json
from PIL import Image
```

```python
cloud= WordCloud(background_color='white',
                 max_words=30,
                 width=400,
                 height=300,
                 font_path='Goyang.otf')

with open('webtoons.json', encoding='utf8') as file:
    webtoons= json.load(file)
    
res= {}
for webtoon in webtoons['webtoons']:
    res[webtoon['title']]= int(float(webtoon['star'])*100)
    
visual= cloud.fit_words(res)
visual.to_image()
visual.to_file('cloud.png')
                 
```

```python
with open('webtoons.json',encoding='utf8') as file:
    webtoons=json.load(file)

res={}
for webtoon in webtoons['webtoons']:
    res[webtoon['title']]= int(float(webtoon['star'])*100)

masking_img= np.array(Image.open('kakao.png'))
cloud= WordCloud(font_path='Goyang.otf',
                 max_font_size=40,
                 mask=masking_img,
                 background_color='white').fit_words(res)
cloud.to_file('result.png')
plt.imshow(cloud,interpolation='bilinear')
plt.axis=('off')
plt.show()
```
