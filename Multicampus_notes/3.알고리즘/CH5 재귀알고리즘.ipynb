```python
#1. 팩토리얼
def factorial(n):
    if n> 0:
        return n* factorial(n-1);
    else:
        return 1
    
n=int(input('정수 n을 입력해주세요 n!값을 구하겠습니다'))
print(f'{n}!값은 {factorial(n)}입니다')
        
```

```python
#2. 최대공약수 함수
def GCD(x,y):
    if x%y ==0:
        return y
    else:
        return GCD(y,x%y)

print(GCD(94,397714))
print(397714/94)
```

```python
def recur(n):
    if n>0:
        recur(n-1)
        print(n)
        recur(n-2)

recur(4)
```

```python
#재귀 알고리즘의 비재귀적 표현
from stack import Stack

def recur(n):
    s= Stack(n)
    
    while True:
        if n>0:
            s.push(n)
            n=n-1
            continue
        if not s.is_empty():
            n=s.pop()
            print(n)
            n=n-2
            continue
        break

recur(4)
# if n>0: 조건문에 따라 s에 4,3,2,1이 저장되고 n=0이된다
# if not s.is_empty() 조건문에 따라 1을 pop하고 출력, n=-1이 된다
# if not s.is_empty() 조건문에 따라 2를 pop하고 출력, n=0이 된다 
# if not s.is_empty() 조건문에 따라 3를 pop하고 출력, n=1이 된다
# if n>0 조건문에 따라 s에 1을 push하여 4,1이 된다, n=0이 된다
# if not s.is_empty() 조건문에 따라 1를 pop하고 출력, n=-1가 된다
# if not s.is_empty() 조건문에 따라 4를 pop하고 출력, n=2가 된다
# if n>0 조건문에 따라 s에 2를 push하여 2가 되고 n=1이 된다
# if n>0 조건문에 따라 s에 1를 push하여 2,1가 되고 n=0이 된다
# if not s.is_empty() 조건문에 따라 1를 pop하고 출력, n=-1가 된다
# if not s.is_empty() 조건문에 따라 2를 pop하고 출력, n=0가 된다
# 출력된 결과: 1>2>3>1>4>1>2
```

```python
#하노이의 탑 구현
def hanoi(n,x,y):
    # 맨 밑바닥의 원반을 제외한 탑 부분을 남은 기둥에 옮기기
    if n>1:
        hanoi(n-1,x,6-x-y)
    print(f'원반[{n}]을(를) {x}기둥에서 {y}기둥으로 옮깁니다')
    
    if n>1:
        hanoi(n-1,6-x-y,y)
        
hanoi(3,1,3)
```

```python
pos= [0]* 8

#pos- list를 출력하는 함수
def put():
    for i in range(8):
        print(f'{pos[i]:2}',end='')
    print()

# i열에 퀸을 배치하라
def pose_queen(i):
    for j in range(8):
        pos[i]= j
        if i== 7:
            put()
        else:
            pose_queen(i+1)

pose_queen(0)
```
